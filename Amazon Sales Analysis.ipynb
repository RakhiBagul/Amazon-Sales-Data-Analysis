# Sales Overview

```python
# Q. What are the monthly and yearly trends in total sales?
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Amount'].sum().reset_index()

# Plot sales trends
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['Month'].astype(str), monthly_sales['Amount'], marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Q How do yearly sales compare?|
```

```python
# Group by year and calculate total sales
yearly_sales = df.groupby('Year')['Amount'].sum().reset_index()

# Plot yearly sales
plt.figure(figsize=(8, 5))
plt.bar(yearly_sales['Year'], yearly_sales['Amount'], color='skyblue')
plt.title('Yearly Sales Comparison')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(yearly_sales['Year'])
plt.show()
```

```python
# Q. 3. Which days of the week generate the most sales?
```

```python
# Extract day of the week
df['Day of Week'] = df['Date'].dt.day_name()

# Group by day of the week
day_sales = df.groupby('Day of Week')['Amount'].sum().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
).reset_index()

# Plot day-wise sales
plt.figure(figsize=(8, 5))
plt.bar(day_sales['Day of Week'], day_sales['Amount'], color='orange')
plt.title('Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Q . 4. How does sales performance vary across different order statuses?
```

```python
# Group by 'Status' and calculate total sales
status_sales = df.groupby('Status')['Amount'].sum().reset_index()

# Plot sales by order status
plt.figure(figsize=(10, 6))
plt.bar(status_sales['Status'], status_sales['Amount'], color='green')
plt.title('Sales by Order Status')
plt.xlabel('Order Status')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
# Q. 5. What is the distribution of sales across different sales channels?/''
```

```python
# Group by 'Sales Channel' and calculate total sales
channel_sales = df.groupby('Sales Channel')['Amount'].sum().reset_index()

# Plot sales by channel
plt.figure(figsize=(8, 5))
plt.bar(channel_sales['Sales Channel'], channel_sales['Amount'], color='purple')
plt.title('Sales by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

# Product Analysis

```python
# Q.  Which product categories and subcategories contribute most to sales?
```

```python
# Group by category and subcategory (replace 'Category' and 'Sub-Category' with actual column names)
product_sales = df.groupby(['Category'])['Amount'].sum().reset_index()

# Sort by sales
product_sales = product_sales.sort_values('Amount', ascending=False)

# Top 10 categories and subcategories
top_products = product_sales.head(10)

# Plot top products
plt.figure(figsize=(12, 6))
plt.barh(top_products['Category'], top_products['Amount'], color='skyblue')
plt.title('Top categories by Amount')
plt.xlabel('Total Sales')
plt.ylabel('Sub-Category')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Q  What are the top 10 product categories by total sales?
```

```python
# Group by 'Category' and calculate total sales
category_sales = df.groupby('Category')['Amount'].sum().reset_index()

# Sort by sales in descending order
top_categories = category_sales.sort_values('Amount', ascending=False).head(10)

# Plot top categories
plt.figure(figsize=(10, 6))
plt.barh(top_categories['Category'], top_categories['Amount'], color='skyblue')
plt.title('Top 10 Product Categories by Sales')
plt.xlabel('Total Sales')
plt.ylabel('Category')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Q. What is the distribution of product sizes?
```

```python
# Group by 'Size' and calculate quantity sold
size_distribution = df.groupby('Size')['Qty'].sum().reset_index()

# Sort sizes by quantity sold
size_distribution = size_distribution.sort_values('Qty', ascending=False)

# Plot size distribution
plt.figure(figsize=(8, 6))
plt.bar(size_distribution['Size'], size_distribution['Qty'], color='orange')
plt.title('Distribution of Product Sizes')
plt.xlabel('Size')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
# Q.  Which product categories have the highest average quantity sold per order?
```

```python
# Calculate average quantity sold per order by category
avg_qty_per_category = df.groupby('Category')['Qty'].mean().reset_index()

# Sort by average quantity in descending order
avg_qty_per_category = avg_qty_per_category.sort_values('Qty', ascending=False)

# Plot average quantity per category
plt.figure(figsize=(10, 6))
plt.bar(avg_qty_per_category['Category'], avg_qty_per_category['Qty'], color='green')
plt.title('Average Quantity Sold per Order by Category')
plt.xlabel('Category')
plt.ylabel('Average Quantity')
plt.xticks(rotation=45)
plt.show()
```

```python
# Q. What is the sales contribution of each product category?
```

```python
# Calculate sales contribution percentage for each category
category_sales['Percentage'] = (category_sales['Amount'] / category_sales['Amount'].sum()) * 100

# Sort by percentage in descending order
category_sales = category_sales.sort_values('Percentage', ascending=False)

# Plot sales contribution
plt.figure(figsize=(10, 6))
plt.pie(category_sales['Percentage'], labels=category_sales['Category'], autopct='%1.1f%%', startangle=140)
plt.title('Sales Contribution by Category')
plt.show()
```

```python
# Q. Which product categories are most frequently sold in different fulfillment methods?
```

```python
# Group by 'Category' and 'Fulfilment' and calculate total quantity sold
fulfillment_category = df.groupby(['Category', 'Fulfilment'])['Qty'].sum().reset_index()

# Sort by quantity sold
fulfillment_category = fulfillment_category.sort_values('Qty', ascending=False)

# Display the top 10 rows
print(fulfillment_category.head(10))

# Plot a pivot table heatmap for visualization (if applicable)
import seaborn as sns
pivot_table = fulfillment_category.pivot(index='Category', columns='Fulfilment', values='Qty')
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='coolwarm')
plt.title('Quantity Sold by Category and Fulfillment Method')
plt.ylabel('Category')
plt.xlabel('Fulfillment Method')
plt.show()
```

# Fulfillment Analysis

```python
# Q. How effective are different fulfillment methods?
```

```python
# Fulfillment method analysis (replace 'Fulfillment Method' with the actual column name)
fulfillment_performance = df.groupby('Fulfilment')['Amount'].sum().reset_index()

# Plot fulfillment methods
plt.figure(figsize=(8, 6))
plt.bar(fulfillment_performance['Fulfilment'], fulfillment_performance['Amount'], color='orange')
plt.title('Sales by Fulfilment')
plt.xlabel('Fulfillment')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Q. What is the average delivery quantity per order for each fulfillment method?
```

```python
# Group by 'Fulfilment' and calculate average quantity
avg_qty_per_fulfillment = df.groupby('Fulfilment')['Qty'].mean().reset_index()

# Sort by average quantity in descending order
avg_qty_per_fulfillment = avg_qty_per_fulfillment.sort_values('Qty', ascending=False)

# Plot average quantity per fulfillment method
plt.figure(figsize=(8, 6))
plt.bar(avg_qty_per_fulfillment['Fulfilment'], avg_qty_per_fulfillment['Qty'], color='orange')
plt.title('Average Quantity per Order by Fulfillment Method')
plt.xlabel('Fulfillment Method')
plt.ylabel('Average Quantity')
plt.xticks(rotation=45)
plt.show()
```

```python
# Q. What is the order status distribution for each fulfillment method?
```

```python
# Group by 'Fulfilment' and 'Status' and count the number of orders
fulfillment_status = df.groupby(['Fulfilment', 'Status'])['Order ID'].count().reset_index()
fulfillment_status.columns = ['Fulfilment', 'Status', 'Order Count']

# Pivot the data for a heatmap
pivot_table = fulfillment_status.pivot(index='Fulfilment', columns='Status', values='Order Count')

# Plot a heatmap
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='Blues')
plt.title('Order Status by Fulfillment Method')
plt.xlabel('Order Status')
plt.ylabel('Fulfillment Method')
plt.show()
```

```python
# Q. What is the relationship between fulfillment methods and sales channels?
```

```python
# Group by 'Fulfilment' and 'Sales Channel' and calculate total sales
fulfillment_channel_sales = df.groupby(['Fulfilment', 'Sales Channel'])['Amount'].sum().reset_index()

# Sort by sales amount
fulfillment_channel_sales = fulfillment_channel_sales.sort_values('Amount', ascending=False)

# Plot a pivot table heatmap
pivot_table = fulfillment_channel_sales.pivot(index='Fulfilment', columns='Sales Channel', values='Amount')

# Visualize using a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='coolwarm')
plt.title('Sales by Fulfillment Method and Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Fulfillment Method')
plt.show()
```

# Customer Segmentation

```python
# top 10 sales
```

```python
# Group by 'ship-city' and 'ship-state' to represent customer location and calculate total sales and number of orders
customer_segments = df.groupby(['ship-city', 'ship-state']).agg({
    'Amount': 'sum',  # Total sales amount
    'Order ID': 'count'  # Number of orders
}).reset_index()

# Rename columns for better readability
customer_segments.rename(columns={
    'Amount': 'Total Sales',
    'Order ID': 'Number of Orders',
    'ship-city': 'City',
    'ship-state': 'State'
}, inplace=True)

# Sort customers by total sales in descending order and display the top 10
top_customers = customer_segments.sort_values('Total Sales', ascending=False).head(10)

# Display the top customers
print("Top 10 Customers by Sales:")
print(top_customers)

# Plot the top 10 customers
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.barh(top_customers['City'] + ', ' + top_customers['State'], top_customers['Total Sales'], color='skyblue')
plt.title('Top 10 Customer Locations by Sales')
plt.xlabel('Total Sales')
plt.ylabel('Customer Location')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Q. What is the average sales per order in each state?
```

```python
# Group by 'ship-state' to calculate the average sales per order
state_avg_sales = df.groupby('ship-state').agg({
    'Amount': 'mean',  # Average sales amount
    'Order ID': 'nunique'  # Total unique orders
}).reset_index()

# Rename columns for clarity
state_avg_sales.rename(columns={'Amount': 'Average Sales per Order', 'Order ID': 'Number of Orders'}, inplace=True)

# Sort by average sales per order in descending order
state_avg_sales = state_avg_sales.sort_values('Average Sales per Order', ascending=False)

# Display the top 10 results
print("Top 10 States by Average Sales per Order:")
print(state_avg_sales.head(10))
```
